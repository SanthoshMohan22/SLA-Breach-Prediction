# Phase 1 : Data Cleaning & Preprocessing

### Step 1: Import Essential Libraries

```python
import pandas as pd
import numpy as np
import warnings
```

### Step 2: Load Dataset

```python
df = pd.read_csv("../data/raw/Customer_support_tickets_Dataset.csv")
print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

### Step 3: Data Inspection

```python
df.info()
```

```python
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values[missing_values > 0])
```

### Step 4: Data Preprocessing

```python
df['started'] = pd.to_datetime(df['started'], errors='coerce')
df['ended'] = pd.to_datetime(df['ended'], errors='coerce')
print("Datetime columns converted successfully")
```

```python
df_cleaned = df.dropna(subset=['started', 'ended'])
print(f"Rows after cleaning: {len(df_cleaned)}")
```

```python
df_cleaned.to_csv("../data/processed/cleaned_tickets.csv", index=False)
print("Cleaned data saved to processed folder")
```
